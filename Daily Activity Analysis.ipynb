```python
## Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
```

```python
df=pd.read_csv("dailyActivity_merged.csv")
```

```python
if "ActivityDate" in df.columns:
    df["ActivityDate"] = pd.to_datetime(df["ActivityDate"])
    df["Day"] = df["ActivityDate"].dt.day_name()
    df["Month"] = df["ActivityDate"].dt.month
```

```python
df
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="TotalSteps", y="Calories", hue="VeryActiveMinutes",palette="viridis",alpha=0.6)
plt.title("Steps vs Calories Burned (colored by Very Active Minutes)")
plt.show()
```

```python
if "SedentaryMinutes" in df.columns:
    df_daily = df.groupby("Day")[["SedentaryMinutes","LightlyActiveMinutes","FairlyActiveMinutes","VeryActiveMinutes"]].mean()
    df_daily = df_daily.loc[["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]]  # reorder days
    
    df_daily.plot(kind="bar", stacked=True, figsize=(10,6), colormap="viridis")
    plt.title("Average Active Minutes per Day")
    plt.ylabel("Minutes")
    plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.histplot(df["TotalSteps"], bins=30, kde=True, color="blue")
plt.title("Distribution of Daily Steps")
plt.xlabel("Steps")
plt.show()
```

```python
activity_cols = ["VeryActiveMinutes","FairlyActiveMinutes","LightlyActiveMinutes","SedentaryMinutes"]

for col in activity_cols:
    plt.figure(figsize=(7,5))
    sns.scatterplot(data=df, x=col, y="Calories", alpha=0.5)
    plt.title(f"{col} vs Calories")
    plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.drop(columns=["ActivityDate","Day"]).corr(), cmap="coolwarm", annot=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
print("""
Key Insights:
1. Steps strongly correlate with calories burned.
2. Sedentary minutes dominate daily time â€” opportunity to encourage more light/active minutes.
3. Very Active Minutes have highest positive effect on calories burned.
4. Weekends often show different patterns â€” possible to target weekend fitness campaigns.
""")
```
